```python
import numpy as np
import matplotlib.pyplot as plt
from lgca import get_lgca
path = '/home/simon/PycharmProjects/invasionlgca/parameter_scans/bigscan/'
betas = np.linspace(0, 10, 51)
ecmdens = np.linspace(0, 10, 51)
N = 5
l = 50
l_spheroid = 2
dims = (l, l)
tmax = 200
restc = 3
rho_0 = 3
nodes = np.zeros((l, l, restc+6), dtype=bool)
nodes[..., :l_spheroid, -rho_0:] = 1
nodes_nt = np.empty((N, tmax, dims[0], dims[1], 6+restc), dtype=bool)
ecm_nt = np.empty((N, tmax, dims[0], dims[1]))

lgca = get_lgca(geometry='hex', interaction='wetting', beta=0., alpha=1, gamma=5, bc='rbc', density=0, restchannels=restc,
                nodes=nodes, rho_0=rho_0)
lgca.r_b = .05
lgca.spheroid = np.zeros_like(lgca.cell_density, dtype=bool)
lgca.spheroid[lgca.r_int:-lgca.r_int, :lgca.r_int+l_spheroid] = 1
lgca.ecm = 1 - lgca.spheroid.astype(float)
lgca.ecm *= 0.
np.savez(path + 'parameters.npz', betas=betas, alpha=lgca.alpha, runs=N, l=l, tmax=tmax, restc=restc, l_spheroid=l_spheroid,
         spheroid=lgca.spheroid, r_b=lgca.r_b, gamma=lgca.gamma, ecmdens=ecmdens)
```

```python
i = 0
for beta in betas:
    lgca.beta = beta
    j = 0
    for ecm in ecmdens:
        print('Progress: beta = {}, ecm = {}'.format(beta, ecm))
        for k in range(N):
            lgca.init_nodes(nodes=nodes)
            lgca.ecm = 1 - lgca.spheroid.astype(float)
            lgca.ecm *= ecm
            lgca.timeevo(tmax//4, recorddens=False, showprogress=False)
            lgca.timeevo(tmax-1, recorddens=False, record=True, showprogress=False, recordecm=True)
            nodes_nt[k] = lgca.nodes_t
            ecm_nt[k] = lgca.ecm_t
        
        j += 1
        np.save(path+'nodes_nt_beta{:.1f}_ecm{:.1f}.npy'.format(beta, ecm), nodes_nt)     
        
    i += 1 
```
